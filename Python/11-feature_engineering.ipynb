````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
pd.options.display.max_columns = None
````

````python
df = pd.read_csv('players_20.csv', index_col = 0)
df.head()
````

````python
df['team_position'].value_counts()
````

````python
starter = df[~df['team_position'].isin(['RES', 'SUB'])]
starter.shape, df.shape
````

So, to begin with, we need to extract a target feature.

````python
starter['team_position']
````

````python
target = starter['team_position'].isin(['DM', 'RW', 'LW', 'LM', 'RM', 'CM', 'AM'])
target.value_counts()
````

### Do a first approximation to use as baseline

````python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, recall_score, precision_score, roc_auc_score, roc_curve
from sklearn.model_selection import train_test_split
````

````python
numeric = starter.select_dtypes('number')
freatures = numeric.fillna(numeric.median())
````

````python
X_train, X_test, y_train, y_test = train_test_split(freatures, target)
````

````python
classifier = LogisticRegression()
classifier.fit(X_train, y_train)
````

````python
accuracy_score(classifier.predict(X_test), y_test)
````

````python
# All in one

numeric = starter.select_dtypes('number')
freatures = numeric.fillna(numeric.median())

X_train, X_test, y_train, y_test = train_test_split(freatures, target)

classifier = LogisticRegression()
classifier.fit(X_train, y_train)
predictions = classifier.predict(X_test)
pred_probs = classifier.predict_proba(X_test)[:,1]

accuracy_score(classifier.predict(X_test), y_test)
tpr, fpr, treshold = roc_curve(y_test, pred_probs)

plt.plot(tpr,fpr)
accuracy_score(y_test, predictions), recall_score(y_test, predictions), precision_score(y_test, predictions), 
roc_curve(y_test, predictions),roc_auc_score(y_test, pred_probs)
````

````python
# All in one - Eliminando 3 variables que est√°n sin escalar en ‚Ç¨

numeric = starter.select_dtypes('number').drop(['wage_eur', 'value_eur', 'release_clause_eur'], axis = 1)
freatures = numeric.fillna(numeric.median())

X_train, X_test, y_train, y_test = train_test_split(freatures, target)

classifier = LogisticRegression()
classifier.fit(X_train, y_train)
predictions = classifier.predict(X_test)
pred_probs = classifier.predict_proba(X_test)[:,1]

accuracy_score(classifier.predict(X_test), y_test)
tpr, fpr, treshold = roc_curve(y_test, pred_probs)

plt.plot(tpr,fpr)
accuracy_score(y_test, predictions), recall_score(y_test, predictions), precision_score(y_test, predictions), 
roc_curve(y_test, predictions),roc_auc_score(y_test, pred_probs)
````

````python
# Empaquetado en funci√≥n y escalando variables

from sklearn.preprocessing import StandardScaler

numeric = starter.select_dtypes('number')
freatures = numeric.fillna(numeric.median())

def evaluate_freatures(freatures, target, classifier = LogisticRegression(), preproc = StandardScaler()):
    
    X_train, X_test, y_train, y_test = train_test_split(freatures, target)
    
    # Train
    X_train = preproc.fit_transform(X_train)
    classifier.fit(X_train, y_train)
    
    #Inference
    X_test = preproc.transform(X_test)
    predictions = classifier.predict(X_test)
    pred_probs = classifier.predict_proba(X_test)[:,1]

    # Evaluation
    tpr, fpr, treshold = roc_curve(y_test, pred_probs)
    plt.plot(tpr,fpr)
    
    return(accuracy_score(y_test, predictions), recall_score(y_test, predictions), precision_score(y_test, predictions), 
    roc_curve(y_test, predictions),roc_auc_score(y_test, pred_probs))

evaluate_freatures(freatures, target)
````

### Target encoding

Also known as bin counting


With high cardinality discrete data like user_ids we would be left with many variables by one hot encoding, most of which would not be very informative. There are [several options](https://www.datacamp.com/community/tutorials/encoding-methodologies) to deal with this. The most sophisticated is probably vector encoding, but we can do with a very simple approach: [target encoding](https://maxhalford.github.io/blog/target-encoding/).
